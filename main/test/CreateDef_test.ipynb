### Parsing the json object

```python
devices_obj = json.loads(devices_json)

for device in devices_obj:
    print(devices_obj[device]['id'])
```

### Saving the json object to file

```python
outputfilename = "device_testbed.json"


with open(outputfilename, 'wb') as outfile:
    outfile.write(devices_json.encode())
    outfile.close()
```
